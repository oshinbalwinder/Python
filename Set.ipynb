# SET
A set is an unordered unique collection of objects in Python

````python
# create a set
s = {"John", "Andy", "Mike", "John", "Amy", "John", "Mike"}
s
````

You can typecast a list into set.

````python
#Convert a list into set
l = [10,20,30,40,50,10,20,70,80]
s1 = set(l)
type(s1)
````

# Set Operations

````python
# Create a Set
A = {"Data Science", "Machine Learning", "Aritifical Intelligence", "Data analytics"}
A
````

### We can add element to the set using ```add``` method 

````python
A.add("Big Data")
A
````

### We can delete an element from set using ```remove``` method

````python
A.remove("Aritifical Intelligence")
A
````

### We can verify whether an element is in set or not using ```in``` command

````python
"Big Data" in A #Returns true if element is present
````

````python
"Artifical Intelligence" in A # returns false if element is not present
````

# Set Logical Operations
1. Intersection
2. Union
3. Difference
4. Superset
5. Subset

````python
# Create two sets
set1 = {10, 20, 30, 40, 50, 60, 70, 80}
set2 = {10, 30, 50, 90, 100, 110}
````

````python
# Print two sets
print(set1, set2)
````

````python
# Intersection
set1.intersection(set2)
````

````python
# Use & to find intersection of two sets
set1&set2
````

````python
# Union
set1.union(set2)
````

````python
#Difference
set1.difference(set2)  # returns elements of set1 that are not present in set2
````

````python
#superset
{10,20,30,40,50}.issuperset({20,30,40})  #checks if {10,20,30,40,50} comprises of all elements of {20,30,40}
````

````python
#subset
{10,20,30}.issubset({10,20,30,40,50}) #Checks if every element of {10,20,30} is in {10,20,30,40,50}
````
